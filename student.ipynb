```python
#checking for outliers
sns.boxplot(x = title_df['runtime_minutes'])
plt.show()
```

```python
# Removing 2 major Outliers
title_df.drop(title_df[title_df['runtime_minutes'] >= 10000].index, inplace = True)
sns.boxplot(x = title_df['runtime_minutes'])
plt.show()
```

```python
#Combining the title to the rating Dataset
title_rating_df = pd.merge(title_df, rating_df, on='tconst')
title_rating_df.head()
```

```python
#changing the data type foreign to Float
gross_df['foreign_gross'] = gross_df['foreign_gross'].str.replace(',', '')
gross_df['foreign_gross'] = gross_df['foreign_gross'].astype(float)
gross_df.dtypes
```

```python
# Adding a calculted column Total Gross
gross_df['total_gross'] = gross_df['domestic_gross'] + gross_df['foreign_gross']
gross_df.head()
```

# Visualizations
The following Visualizations have been used:
-  Bar chart
-  Scatter plot
-  Line graph
-  Horizontal Bar chart
-  Histogram

```python
# bar chart best rated genre

genre_ratings = title_rating_df.groupby('genres')['averagerating'].mean().reset_index()

# Sort the genres by rating in descending order and select the top 5
top_5_genres = genre_ratings.nlargest(5, 'averagerating')

# Create a bar graph to visualize the top 5 highest-rated genres
plt.figure(figsize=(9, 7))
plt.bar(top_5_genres['genres'], top_5_genres['averagerating'])
plt.xlabel('Movie Genre')
plt.ylabel('Average Rating')
plt.title('Top 5 Highest-Rated Movie Genres')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.tight_layout()

# Show the bar graph
plt.show()
```

```python
# Scatter plot best rated genre
genre_ratings = title_rating_df.groupby('genres')['averagerating'].mean().reset_index()

# Sort the genres by rating in descending order and select the top 5
top_5_genres = genre_ratings.nlargest(5, 'averagerating')

# Create a scatter plot to visualize the top 5 highest-rated genres
plt.figure(figsize=(9, 7))
plt.scatter(top_5_genres['genres'], top_5_genres['averagerating'])
plt.xlabel('Movie Genre')
plt.ylabel('Average Rating')
plt.title('Top 5 Highest-Rated Movie Genres')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.tight_layout()

# Show the scatter plot
plt.show()
```

```python
#Bar chart highest studio in total gross
studio_ratings = gross_df.groupby('studio')['total_gross'].sum().reset_index()

# Sort the studios by rating in descending order and select the top 5
top_5_studio = studio_ratings.nlargest(5, 'total_gross')

# Create a bar graph to visualize the top 5 studios with highest total gross
plt.figure(figsize=(9, 7))
plt.bar(top_5_studio['studio'], top_5_studio['total_gross'], color='green')
plt.xlabel('studios')
plt.ylabel('total gross')
plt.title('Top 5 studio with Highest total gross')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.tight_layout()

# Show the bar graph
plt.show()
```

```python
#Line graph highest studio in total gross
studio_ratings = gross_df.groupby('studio')['total_gross'].sum().reset_index()

# Sort the studios by rating in descending order and select the top 5
top_5_studio = studio_ratings.nlargest(5, 'total_gross')

# Create a bar graph to visualize the top 5 studios with highest total gross
plt.figure(figsize=(9, 7))
plt.plot(top_5_studio['studio'], top_5_studio['total_gross'], color='green')
plt.xlabel('studios')
plt.ylabel('total gross')
plt.title('Top 5 Highest-Rated Movie Genres')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.tight_layout()

# Show the bar graph
plt.show()
```

```python
#Revoming null values in runtime_minutes column
title_rating_df.dropna()
```

```python
#Horizontal bar graph Highest-Rated Movie Genres with Average Runtime
genre_runtime = title_rating_df.groupby('genres')['runtime_minutes'].mean().reset_index()
genre_ratings_runtime = genre_ratings.merge(genre_runtime, on='genres', how='inner')

# Sort the genres by rating in descending order and select the top 5
top_5_genres = genre_ratings_runtime.nlargest(5, 'averagerating')

# Create a bar graph to visualize the top 5 highest-rated genres
plt.figure(figsize=(9, 7))
plt.barh(top_5_genres['genres'], top_5_genres['averagerating'], label='Average Rating')
plt.barh(top_5_genres['genres'], top_5_genres['runtime_minutes'], label='Average Runtime (minutes)')
#plt.bar(genre_runtime['genres'], genre_runtime['runtime_minutes'])
plt.xlabel('Average Rating / Average Runtime (minutes)')
plt.ylabel('Movie Genre')
plt.title('Top 5 Highest-Rated Movie Genres with Average Runtime')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.tight_layout()

plt.show()
```

```python
#Histogram Average Ratings by Genre
genre_ratings = title_rating_df.groupby('genres')['averagerating'].mean().reset_index()

# Create a histogram of the average ratings for each genre
plt.figure(figsize=(10, 6))
plt.hist(genre_ratings['averagerating'], bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('averagerating')
plt.ylabel('Frequency')
plt.title('Histogram of Average Ratings by Genre')
plt.grid(True)

# Show the histogram
plt.show()

```
